```python
%%writefile HW7.cpp
//10-7

#include <iostream>
#include <iomanip>
#include <string>
#include <string.h>
using namespace std;

class SavingsAccount {
    //member function
public:
    SavingsAccount(); //constructor

    float calculateMonthlyInterest(){
        //calculate and assign the value to saving balance
        monthlyInterest = this->getBalance() * (annualInterestRate/12);
        savingsBalance += monthlyInterest;
    }

    static void modifyInterestRate(float aIR){
        //set rate to new value
        annualInterestRate = aIR;
    }

    void setBalance(float bb){
        savingsBalance = bb;
    }

    float getBalance(){
        return savingsBalance;   
    }

    void printBalance(){
        cout << "The balance is: " << savingsBalance <<endl;
    }

    //data member
private:
    static float annualInterestRate;
    float savingsBalance;
    float monthlyInterest;

};  //end class SavingsAccount

//constructor
SavingsAccount::SavingsAccount(){
        savingsBalance = monthlyInterest = 0.0;

}

//static的member值的宣告位置必須在 global namespace scope!!!!
float SavingsAccount::annualInterestRate = 0.0;

int main(){
    SavingsAccount saver1;
    SavingsAccount saver2; 

    cout<< "default set. saver1: 2000, saver2: 3000\n" << endl;
    saver1.setBalance(2000.0);
    saver2.setBalance(3000.0);

//1 reset the annual interest
    cout<< "#1 set annual interest rate as 0.03" << endl;
    saver1.modifyInterestRate(0.03);

    saver1.calculateMonthlyInterest();
    saver2.calculateMonthlyInterest();

    saver1.printBalance();
    saver2.printBalance();
    cout << "\n";

//2 reset the annual interest
    cout<< "#2 set annual interest rate as 0.03" << endl;
    saver1.modifyInterestRate(0.04);

    saver1.calculateMonthlyInterest();
    saver2.calculateMonthlyInterest();

    saver1.printBalance();
    saver2.printBalance();
    cout << "\n";

    return 0;
}
```
