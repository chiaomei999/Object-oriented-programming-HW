```python
%%writefile HW2_1.cpp

#include <iostream>
#include <iomanip>
using namespace std;

//給定一種新的type叫做Account
class Account{
    //member funct.
    public:
      Account(); //constructor (default值，不能return任何東西)
      void setBalance(int);
      void Credit(int);
      void Debit(int); 
      void getBalance();

    //data member  //data members are normally private
    private:
      int Balance;
}; //end class Account

//Account Constructor
Account::Account(){
    Balance = 0;
} //end Account Constructor

void Account::setBalance(int aBalance){  //validate balance
    if( aBalance >= 0 ){
        Balance = aBalance;
    }
    else{
        Balance = 0;
    }
} //end funct. setBalance

void Account::Credit(int add){    
  Balance += add;
} //end fuct.

void Account::Debit(int withdraw){    
  if( withdraw <= Balance ){
      Balance -= withdraw;
  }
  else{
      cout << "Debit amount exceeded account balance. Cancel this action." <<endl ;
  }
} //end fuct.

void Account::getBalance(){    
  cout << Balance;
} //end fuct.


int main(){

    //MattAccount 有個type Account的物件叫MattAccount
    Account MattAccount;

    //給定值
    MattAccount.setBalance(80);  //change Balance
    MattAccount.Credit(300);
    MattAccount.Debit(1500);

    cout << "MattAccount's Balance is: ";
    MattAccount.getBalance();  //用這個來cout Balance

    //MacieAccount 有個type Account的物件叫MacieAccount
    Account MacieAccount;
    
    //給定值
    MacieAccount.Credit(3000);
    MacieAccount.Debit(1277);

    cout << "\nMacieAccount's Balance is: ";
    MacieAccount.getBalance();  //用這個來cout Balance
    return 0;
}
```

```python
%%writefile HW2_2.cpp

#include <iostream>
#include <iomanip>
#include <string>
#include <string.h>
using namespace std;

//給定一種新的type叫做Invoice
class Invoice{
    //member function
    public:
      Invoice(); //constructor (default值，不能return任何東西)
      void Set(string,string,int,int);
      void Get();
      void getInvoiceAmount(); 

    //data member
    private:
      string number;
      string description;
      int quantity;
      int price;
}; //end class Invoice

//Invoice Constructor
Invoice::Invoice(){
    number = "20211224";
    description = "ProductName";
    quantity = 0;
    price = 0;
} //end Invoice Constructor

void Invoice::Set(string numberA, string descriptionA, int quantityA, int priceA){
    number = numberA;
    description = descriptionA;
    quantity = (quantityA >=0 )?quantityA :0;  //validate
    price = (priceA >=0 )?priceA :0;      //validate
} //end funct.


void Invoice::Get(){    
  cout << "- number: " << number <<endl;
  cout << "- description: " << description <<endl;
  cout << "- quantity: " << quantity <<endl;
  cout << "- price: " << price <<endl;
} //end funct.

void Invoice::getInvoiceAmount(){    
  cout << price*quantity;
} //end fuct.


int main(){

    //MyInvoice 有個type Invoice的物件叫MyInvoice
    Invoice MyInvoice;

    //給定值
    MyInvoice.Set("20220402","100-day gift",87,2);  //change content

    cout << "MyInvoice's content is: " << endl;
    MyInvoice.Get();  //要記得加() 不然沒辦法run

    cout << "MyInvoice's amount is: ";
    MyInvoice.getInvoiceAmount();  //用這個來cout

    return 0;
}
```
