```python
%%writefile HW9.cpp

#include <iostream>
#include <iomanip>
#include <string>
#include <string.h>
using namespace std;

//base class

class Package{
//member function
public:
    //constructor
    Package(string, string, string, string, int, double, double); 
    double calculateCost(){
        totalCost = weight * cost;
        return totalCost;
    };

    void setWeight(double);
    void setCost(double);
    double getWeight();
    double getCost();

//data member
protected:
    string name;
    string address;
    string city;
    string state;
    int zip;
    double weight; //>=0
    double cost; //>=0
    double totalCost;

};


//constructor
Package::Package(string nn, string aa, string cc, string ss, int zz, double ww, double co){
    name = nn;
    address = aa;
    city = cc;
    state = ss;
    zip = zz;
    setWeight(ww) ; //validate >=0
    setCost(co);   //validate >=0
}

void Package::setWeight(double ww){
    weight = (ww >= 0.0) ?ww :0.0;
}

void Package::setCost(double co){
    cost = (co >= 0.0) ?co :0.0;
}

double Package::getWeight(){
    return weight;
}

double Package::getCost(){
    return cost;
}

//-----------  class 1: TwoDayPackage ---------------

class TwoDayPackage : public Package{
public:
  TwoDayPackage(string, string, string, string, int, double, double, double);

  double calculateCost(){
    totalCost = weight * (cost + flatfee);
    return totalCost;
  };

  double setFlatFee(double fl){
    flatfee = fl;
  };

  double getFlatFee(){
    return flatfee;
  };

private:
  double flatfee;
};

//constructor
TwoDayPackage::TwoDayPackage(string nn, string aa, string cc, string ss, int zz, double ww, double co, double fl)
    //call base constructor 
    : Package(nn, aa, cc, ss, zz, ww, co){  //type
      setFlatFee(fl);
    }

//-----------  class 2: OverNightPackage ----------

class OverNightPackage : public Package{
public:
  OverNightPackage(string, string, string, string, int, double, double, double);

  double calculateCost(){
    totalCost = weight * (cost + addfee);
    return totalCost;
  };

  double setAddFee(double af){
    addfee = af;
  };

  double getAddFee(){
    return addfee;
  };

private:
  double addfee;
};

//constructor
OverNightPackage::OverNightPackage(string nn, string aa, string cc, string ss, int zz, double ww, double co, double af)
    //call base constructor 
    : Package(nn, aa, cc, ss, zz, ww, co){
      setAddFee(af);
    }

//------------ main -----------------

int main(){
    
    cout<< "Base Class:" << endl;
    Package P1("Matt", "my road no.1", "Tainan", "Taiwan", 777, 10.3, 5);
    cout<< "P1: cost, weight: " << P1.getCost() << " , " << P1.getWeight() << endl;
    cout << "P1 total cost is: " << P1.calculateCost() << endl << endl;

    cout<< " Class 1:" << endl;
    TwoDayPackage P2("Macie", "my road no.2", "Tainan", "Taiwan", 888, 10.3, 5, 0.4);
    cout<< "P2: cost, weight, flatfee: " << P2.getCost() << " , " << P2.getWeight() <<  " , " << P2.getFlatFee() << endl;
    cout << "P2 total cost is: " << P2.calculateCost() << endl << endl;

    cout<< " Class 2:" << endl;
    OverNightPackage P3("Ania", "my road no.3", "Spy", "Taiwan", 999, 10.3, 5, 0.8);
    cout<< "P3: cost, weight, addfee: " << P3.getCost() << " , " << P3.getWeight() <<  " , " << P3.getAddFee() << endl;
    cout << "P3 total cost is: " << P3.calculateCost() << endl;


    return 0;
}

```
